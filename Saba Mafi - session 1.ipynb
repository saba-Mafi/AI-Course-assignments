```python
import numpy as np
from numpy import random
import time
x=random.randint(10000,size=(500000))
print(x)

start_time_with_numpy=time.time()
max_num_withnumpy=x.max()
finish_time_with_numpy=time.time()
print("maximum with numpy function: " + str(max_num_withnumpy))
total_time_with_numpy= finish_time_with_numpy-start_time_with_numpy

print("total execution time using numpy is : " + str(total_time_with_numpy))

start_time_with_forloop=time.time()
max_num_withforloop=x[0]
for items in x:
    if items>max_num_withforloop:
        max_num_withforloop=items
finish_time_with_forloop=time.time()
total_time_with_forloop=finish_time_with_forloop-start_time_with_forloop

print ("maximum with for loop:  "+str(max_num_withforloop))
print("total execution time using for loop is : " + str(total_time_with_forloop))

```
